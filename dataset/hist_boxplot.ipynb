```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset("iris")
```

```python
print(iris)
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
sns.histplot(iris["petal_length"],bins=20,kde=True,color="lightgreen",edgecolor="black")
plt.title("histogram petal length")
plt.xlabel("petal length")
plt.ylabel("frequency")

plt.subplot(1,2,2)
sns.boxplot(x="petal_length",data=iris,color="lightblue")
plt.title("boxplot petal length")
plt.xlabel("species")
plt.ylabel("petal_length")
plt.tight_layout()
plt.show()  
```
